```python
import folium
import pandas as pd
import numpy as np

%matplotlib inline
```

# Grafica en mapa

```python
df = pd.read_csv('exa__dispositivos_cali.csv')
data = pd.DataFrame(df[:50], columns = ['latitud','longitud']) #[:50]  <-this kind of slice is useful for developing a map
data.head()
```

```python
# se agreag una columna de color 
def colores(row):
    if row['latitud'] > 3.43:
        return 'blue'
    else:
        return 'red'

data['color'] = data.apply(colores, axis=1)
data.head()
```

```python
map_osm = folium.Map(location=[3.4, -76.5], zoom_start=11)

data.apply(lambda row:folium.CircleMarker(location=[row["latitud"], row["longitud"]], 
                                              radius=3, fill_color=row['color'], color=row['color'])
                                             .add_to(map_osm), axis=1)

map_osm
```
